# LABEL ENCODER

```python
df = df.apply(LabelEncoder().fit_transform)
```

```python
plt.figure(figsize=[15,80])
j = 0
for column,dtype in df.dtypes.items():
    if dtype != 'object':
        j += 1
        plt.subplot(20,2,j)
        sns.boxplot(df[column])
```


We can see that
There is Lots of Outliers in salary column
and few outliers in Number of company workerd,Performance rating, Stock Option Level,TotalWorking Years,Traning time last year,Years at company,Yearsin Current Role, Years ScienceLast Promotation, Years With Curr Manager #### So removing outliers

```python
abs_zscore = np.abs(stats.zscore(df))
abs_zscore
```

```python
df[abs_zscore > 3].index
```

```python
df.drop(df[abs_zscore > 3].index,axis=0,inplace=True)
```

```python
df['Attrition'].value_counts()
```

# SPLITING VARIABLE

```python

X = df.drop('Attrition',axis=1)
```

```python
y = df['Attrition']
```

# CHECKING VIF

```python
df_info = pd.DataFrame()
df_info['VIF'] = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
df_info['columns'] = X.columns
df_info.sort_values('VIF',ascending=False)
```

There is very low Multi-colierity in Data so leaving it¶

# Checking for skewness

```python
df.skew().abs().sort_values()
```

There is a skewness in data so doing PowerTransformation

```python
X_transformed = PowerTransformer().fit_transform(X)
X = pd.DataFrame(X_transformed,columns=X.columns)
X
```

# TRAIN TEST SPLIT

```python
X_train, X_test,y_train,y_test = train_test_split(X,y,random_state=42,shuffle=True)
```

MODEL BUILDING

# LOGISTIC REGRESSION

```python
logisticRegression = LogisticRegression()
```

```python
logisticRegression.fit(X_train,y_train)
```

```python

y_pred = logisticRegression.predict(X_test)
y_pred_proba = logisticRegression.predict_proba(X_test)[::,1]
```

```python
metrics.accuracy_score(y_test,y_pred)
```

```python
metrics.confusion_matrix(y_test,y_pred)
```

```python
fpr, tpr , _ = metrics.roc_curve(y_test,y_pred_proba)
```

```python
auc = metrics.roc_auc_score(y_test,y_pred_proba)
print('auc:',auc)
```

```python
plt.plot(fpr,tpr)
```

```python
kFold = KFold(n_splits=10,shuffle=True,random_state=42)
np.mean(cross_val_score(LogisticRegression(),X,y,cv=kFold,scoring='accuracy',n_jobs=1))
```
